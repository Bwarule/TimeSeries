```python
def stock_perfo(stock_data):
    ## current price
    current_value = stock_data.loc[stock_data.index == stock_data.index[stock_data.shape[0]-1],'Close']
    current_value
    ## Last 5days price
    Last5days = stock_data.loc[stock_data.index == stock_data.index[stock_data.shape[0]-5],'Close']
    Last5days
    ## last 5days return 
    print(current_value[0],Last5days[0])
    R5 = (current_value[0]-Last5days[0])/current_value[0]
    R5

    Last2w = stock_data.loc[stock_data.index == stock_data.index[stock_data.shape[0]-10],'Close']
    print('Current price', current_value[0],'; Last 2 week',Last2w[0])
    R_Last2w = (current_value[0]-Last2w[0])/current_value[0]
    R_Last2w

    Last3w = stock_data.loc[stock_data.index == stock_data.index[stock_data.shape[0]-15],'Close']
    print('Current price', current_value[0],'; Last 3 week',Last3w[0])
    R_Last3w = (Last2w[0]-Last3w[0])/Last2w[0]
    R_Last3w
    
    Last4w = stock_data.loc[stock_data.index == stock_data.index[stock_data.shape[0]-20],'Close']
    print('Current price', current_value[0],'; Last 4 week',Last4w[0])
    R_Last4w = (Last3w[0]-Last4w[0])/Last3w[0]
    R_Last4w
    
    Last2m = stock_data.loc[stock_data.index == stock_data.index[stock_data.shape[0]-40],'Close']
    print('Current price', current_value[0],'; Last 2 Months',Last2m[0])
    R_Last2m = (Last4w[0]-Last2m[0])/Last4w[0]
    R_Last2m
    
    Last3m = stock_data.loc[stock_data.index == stock_data.index[stock_data.shape[0]-65],'Close']
    print('Current price', current_value[0],'; Last 3 Months',Last3m[0])
    R_Last3m = (Last2m[0]-Last3m[0])/Last2m[0]
    R_Last3m
    
      
    # Last 1w,2w,3w,4w,2m,3m
    dict = {'Symbol':[stock_data['Symbol'][0]], '1w': [R5], '2w' : [R_Last2w], '3w' : [R_Last3w],
            '4w' : [R_Last4w], '2m':[R_Last2m] ,'3m':[R_Last3m]}
    
    dict1 = {'Symbol':[stock_data['Symbol'][0]],
            'current_value':[current_value[0]],
            '1w': [Last5days[0]],
            '2w' : [Last2w[0]], 
            '3w' : [Last3w[0]],
            '4w' : [Last4w[0]],
             '2m':[Last2m[0]] ,'3m':[Last3m[0]]}
    # creating a dataframe from dictionar
    df_output = pd.DataFrame(dict)
    
    return  df_output
    
```

```python
def stock_perfo_byw(stock_data):
    ## current price
    current_value = stock_data.loc[stock_data.index == stock_data.index[stock_data.shape[0]-1],'Close'][0]
    
    D1 = stock_data.loc[stock_data.index == stock_data.index[stock_data.shape[0]-1],'Close'][0]
    D2 = stock_data.loc[stock_data.index == stock_data.index[stock_data.shape[0]-2],'Close'][0]
    D3 = stock_data.loc[stock_data.index == stock_data.index[stock_data.shape[0]-3],'Close'][0]
    D4 = stock_data.loc[stock_data.index == stock_data.index[stock_data.shape[0]-4],'Close'][0]
    D5 = stock_data.loc[stock_data.index == stock_data.index[stock_data.shape[0]-5],'Close'][0]
    D6 = stock_data.loc[stock_data.index == stock_data.index[stock_data.shape[0]-6],'Close'][0]
    D7 = stock_data.loc[stock_data.index == stock_data.index[stock_data.shape[0]-7],'Close'][0]
    D8 = stock_data.loc[stock_data.index == stock_data.index[stock_data.shape[0]-8],'Close'][0]
    D9 = stock_data.loc[stock_data.index == stock_data.index[stock_data.shape[0]-9],'Close'][0]
    D10 = stock_data.loc[stock_data.index == stock_data.index[stock_data.shape[0]-10],'Close'][0]
    D11 = stock_data.loc[stock_data.index == stock_data.index[stock_data.shape[0]-11],'Close'][0]
    D12 = stock_data.loc[stock_data.index == stock_data.index[stock_data.shape[0]-12],'Close'][0]
    
    ## Last 5days price
    w1 = stock_data.loc[stock_data.index == stock_data.index[stock_data.shape[0]-6],'Close'][0]
    
    w2 = stock_data.loc[stock_data.index == stock_data.index[stock_data.shape[0]-11],'Close'][0]
    
    w3 = stock_data.loc[stock_data.index == stock_data.index[stock_data.shape[0]-16],'Close'][0]
    
    w4 = stock_data.loc[stock_data.index == stock_data.index[stock_data.shape[0]-21],'Close'][0]
    
    w5 = stock_data.loc[stock_data.index == stock_data.index[stock_data.shape[0]-26],'Close'][0]
    
    w6 = stock_data.loc[stock_data.index == stock_data.index[stock_data.shape[0]-31],'Close'][0]
    
    w7 = stock_data.loc[stock_data.index == stock_data.index[stock_data.shape[0]-36],'Close'][0]
    
    w8 = stock_data.loc[stock_data.index == stock_data.index[stock_data.shape[0]-41],'Close'][0]
    
    w9 = stock_data.loc[stock_data.index == stock_data.index[stock_data.shape[0]-46],'Close'][0]
    
    w10 = stock_data.loc[stock_data.index == stock_data.index[stock_data.shape[0]-51],'Close'][0]
    
    ## last 5days return 
          
    # Last 1w,2w,3w,4w,2m,3m
    dict = {'Symbol':[stock_data['Symbol'][0]], 
            #'1D' :[(current_value-D1)/D1],
            '2D' :[(D1-D2)/D2],  '3D' : [(D2-D3)/D3], '4D' :[(D3-D4)/D4],  '5D': [(D4-D5)/D5],
            '6D' :[(D5-D6)/D6],  '7D' : [(D6-D7)/D7], '8D' :[(D7-D8)/D8],  '9D': [(D8-D9)/D9],
            '10D': [(D9-D10)/D10], '11D': [(D10-D11)/D11],'12D': [(D11-D12)/D12],
            'Week_wise' : [''],
            '1w': [(current_value-w1)/w1], '2w' : [(w1-w2)/w2], 
            '3w' : [(w2-w3)/w3], '4w' : [(w3-w4)/w4], '5w': [(w4-w5)/w5], '6w' : [(w5-w6)/w6],
            '7w' : [(w6-w7)/w7], '8w' : [(w7-w8)/w8], '9w': [(w8-w9)/w9], '10w': [(w9-w10)/w10]
            }
    
    # creating a dataframe from dictionar
    df_output = pd.DataFrame(dict)
    
    return  df_output
    
```
